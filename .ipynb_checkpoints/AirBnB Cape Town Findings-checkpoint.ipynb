Here I decided to drop some of the columns that don't offer us any valuable information.

*After note: I decided to drop the 'neighbourhoof_cleansed' field as there is too little data points for each category and it leads to bad performance from the model*

**Helper functions:**

```python
def percentage_to_rate(percent):
    '''
    Converts a string percentage to a float value.
    
    INPUT:
        percent - String representation of a percentage
        
    OUTPUT:
        float - Float representation of a percentage ranging from 0 to 1.0
    '''
    
    if (isinstance(percent, float)):
        return percent
    
    return float(percent.split("%")[0]) / 100

def airbnb_bool_to_bool(airbnb_bool):
    '''
    Converts Airbnb's representation of a boolean to Python boolean
    
    INPUT:
        airbnb_bool - String (either t or f)
        
    OUTPUT:
        boolean - Python boolean
    '''
    
    if (isinstance(airbnb_bool, int)):
        return airbnb_bool

    
    return 1 if airbnb_bool == "t" else 0

def months_since(date_str):
    '''
    Calculates the months between today's date and the date entered
    
    INPUT:
        date_str - The date represented in string format (2023-12-09)
        
    OUTPUT:
        number - The number of months since the given date
    '''
    
    if (isinstance(date_str, int)): 
        return date_str
    
    years = int(date.today().year) - int(date_str.rsplit("-")[0])
    months = int(date.today().month) - int(date_str.rsplit("-")[1])
    
    total_months = (years * 12) + months

    return total_months if total_months >= 0 else 0

def price_to_float(price):
    '''
    Converts a string representation of a price in dollars to a float value
    
    INPUT:
        price - String representation of a price ($200)
        
    OUTPUT:
        float - Value of the price
    '''
    
    if (isinstance(price, float)): 
        return price
    
    return float(price.replace(",", "").replace("$", ""))
```

Here we are standardizing the input data. 

We are changing all the columns that represent percentages as a string into a float value for better model adaption.

All the booloean columns are currently represented with t for true and f for false; so we are converting those into python boolean representations.

I have also added new fields to keep track of whether listings have a description or not (this process has been carried over for other columns such as neighbourhood_overview, location, about and profile_pic).

Another column I've added is to calculate the months a host has had their listing and it has been active.

Lastly I converted the string representation of the price to a float value so it performs as a number feature.

```python
df_reduced['listing_has_description'] = df_reduced['description'].notnull().replace({True: 1, False: 0})
df_reduced['listing_has_neighborhood_overview'] = df_reduced['neighborhood_overview'].notnull().replace({True: 1, False: 0})
df_reduced['host_months_on_airbnb'] = df_reduced['host_since'].map(months_since, na_action="ignore")
df_reduced['host_has_location'] = df_reduced['host_location'].notnull().replace({True: 1, False: 0})
df_reduced['host_has_about'] = df_reduced['host_about'].notnull().replace({True: 1, False: 0})
df_reduced['host_acceptance_rate'] = df_reduced['host_acceptance_rate'].map(percentage_to_rate, na_action="ignore")
df_reduced['host_is_superhost'] = df_reduced['host_is_superhost'].map(airbnb_bool_to_bool, na_action="ignore")
df_reduced['instant_bookable'] = df_reduced['instant_bookable'].map(airbnb_bool_to_bool, na_action="ignore")
df_reduced['host_has_profile_pic'] = df_reduced['host_has_profile_pic'].map(airbnb_bool_to_bool, na_action="ignore")
df_reduced['host_identity_verified'] = df_reduced['host_identity_verified'].map(airbnb_bool_to_bool, na_action="ignore")
df_reduced['price'] = df_reduced['price'].map(price_to_float, na_action="ignore")
df_reduced['months_since_last_review'] = df_reduced['last_review'].map(months_since, na_action="ignore")
df_reduced['months_since_first_review'] = df_reduced['first_review'].map(months_since, na_action="ignore")
df_reduced['listing_months_active'] = df_reduced['months_since_first_review']
```

### I will be using df_reduced for the rest of the notebook and treat it as the input data

Below I have created two functions to handle the rest of the cleaning process for the (now standardized) dataset.

The cleaning process procedes as follows:
1. Drop all the records that has a NaN value for the target column
2. Split the target column from the features dataframe
3. We fill all the empty values for numeric columns with the mean and in fields where that doesn't make sense we use 0, minimum value or the maximum value
4. Clean the amenities (This involves finding all the types of amenities, sorting it from most to least in how many listings offer them and finally creating dummy columns to represent if a listing has an amenity or not)
5. To deal with the rest of the categorical columns, we create dummy columns using pandas built-in function
6. Drop any new columns created that have the same value for all records

```python
def clean_amenities(df, max_amenities = 0):
    '''
    Cleans the ameneties column but filtering out the ameneties and creating new columns to indicate if a listing offers the amenity.
    
    INPUT
    df - pandas dataframe
    max_amenities - The top number of amenities to add as dummy columns to the dataframe
    
    OUTPUT
    amenities_dict_limitted - A dictionary containing the top x amenities with their name and the number of occurrences
    '''
    if max_amenities == 0:
        df.drop('amenities', inplace=True, axis=1)
        return {}
    
    amenities_dict = {}

    for row in df['amenities']:
        options = row.strip('][').replace('"', '').split(', ')
    
        for amenity in options:
            amenity_name = "amenities_" + amenity.replace(" ", "_")
        
            if amenity_name in amenities_dict:
                amenities_dict[amenity_name] += 1
            else:
                amenities_dict[amenity_name] = 1
    
    amenities_list = sorted(amenities_dict.items(), key=lambda x:x[1], reverse=True)
    
    amenities_dict_limitted = dict(amenities_list[:max_amenities])
    
    df.reset_index(inplace=True, drop=True)

    for amenities_name in amenities_dict_limitted: 
        df[amenities_name] = 0
        
    for i in range(df.shape[0]):
        am = df.at[i, 'amenities']
        options = am.strip('][').replace('"', '').split(', ')
    
        for amenity in options:
            amenity_name = "amenities_" + amenity.replace(" ", "_")
            
            if amenity_name in amenities_dict_limitted:
                df.at[i, amenity_name] = 1
                
    df.drop('amenities', inplace=True, axis=1)
                
    return amenities_dict_limitted
    
```

```python
def clean_data(df, target_column, max_amenities = 365):
    '''
    Cleans the standardized dataset and splits the data into the features and the target dataframe.
    
    INPUT
    df - pandas dataframe
    target_column - Column being used as the target column (column to predict)
    max_amenities - The top x amount of amenities to keep
    
    OUTPUT
    X - A matrix holding all of the variables you want to consider when predicting the response
    y - the corresponding response vector
    '''
    # Drop rows with missing salary values
    df.dropna(subset=[target_column], axis=0, inplace=True)
    
    y = df[target_column]
    df = df.drop(target_column, axis=1)
    
    # Fill numeric values
    df['host_acceptance_rate'].fillna(df['host_acceptance_rate'].mean(), inplace=True)
    df['host_is_superhost'].fillna(0, inplace=True)
    df['bedrooms'].fillna(0, inplace=True)
    df['beds'].fillna(0, inplace=True)
    df['minimum_nights_avg_ntm'].fillna(1, inplace=True)
    df['maximum_nights_avg_ntm'].fillna(3000, inplace=True)
    
    for col in ['review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 'review_scores_value']:
        if 'review_scores_rating' in df.columns:
            df[col].fillna(df['review_scores_rating'], inplace=True)
        else:
            df[col].fillna(df[col].mean(), inplace=True)
        
    # Deal with amenities
    amenities_dict = clean_amenities(df, max_amenities)
    
    df = pd.get_dummies(df, columns=df.select_dtypes(include=['object']).columns, dummy_na=True)
    
    if 'room_type_nan' in df.columns:
        df.drop(['room_type_nan'], axis=1, inplace=True)
    
    X = df
    return X, y
    
#Use the function to create X and y
X, y = clean_data(df_reduced, 'reviews_per_month', 100)  
```

## Data Modeling

### Model Fitting

```python
def create_model(X, y):
    '''
    Creates a linear predictive model from the given inputs
    
    INPUT:
        X: pd.DataFrame - Dataframe containing all the features
        y: pd.Series - The target column
    
    OUTPUT:
        linear_model - The fitted model from the data
        train_score - The R2 Score from the training data
        test_score - The R2 Score from the testing data
        X_train - The training data based on the train/test split
    '''
        
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .30, random_state=42)
    
    linear_model = LinearRegression(normalize=True)
    linear_model.fit(X_train, y_train)
    
    y_test_preds = linear_model.predict(X_test)
    y_train_preds = linear_model.predict(X_train)
    
    test_score = r2_score(y_test, y_test_preds)
    train_score = r2_score(y_train, y_train_preds)
    
    return linear_model, train_score, test_score, X_train
```

```python
linear_model, train_score, test_score, X_train = create_model(X, y)
```

```python
print("The rsquared on the training data was {}.  The rsquared on the test data was {}.".format(train_score, test_score))
```

```python
def coef_weights(coefficients, X_train):
    '''
    INPUT:
    coefficients - the coefficients of the linear model 
    X_train - the training data, so the column names can be used
    OUTPUT:
    coefs_df - a dataframe holding the coefficient, estimate, and abs(estimate)
    
    Provides a dataframe that can be used to understand the most influential coefficients
    in a linear model by providing the coefficient estimates along with the name of the 
    variable attached to the coefficient.
    '''
    coefs_df = pd.DataFrame()
    coefs_df['est_int'] = X_train.columns
    coefs_df['coefs'] = coefficients
    coefs_df['abs_coefs'] = np.abs(coefficients)
    coefs_df = coefs_df.sort_values('abs_coefs', ascending=False)
    return coefs_df

#Use the function
coef_df = coef_weights(linear_model.coef_, X_train)

#A quick look at the top results
coef_df.head(20)
```

We are using the coefficient weights to determine the feature importance which can help determine what makes a good listing.

Something that should be noted is that the room_types and host_response_time columns all have the same coefficient weights. This could indicate that these columns don't have any effect on the reviews_per_month and I will explore dropping these columns later on in the notebook.

### What amount of amenities leads to the best model?

I want to see what the effect of the amount of ameneties we include in the training data effect the overall performance of the model

```python
num_amenities = [0, 1, 3, 5, 10, 50, 100, 150, 200]

train_scores = []
test_scores = []

for num in num_amenities:
    X, y = clean_data(df_reduced, 'reviews_per_month', num)
    lm_model, train_score, test_score, X_train = create_model(X, y)
    
    train_scores.append(round(train_score, 2))
    test_scores.append(round(test_score, 2))
    
plt.plot(train_scores, num_amenities, label='Training Scores')
plt.plot(test_scores, num_amenities, label='Testing Scores')
plt.legend()
plt.show()
```

The graph seems to indicate increasing the amount of amenities has no direct effect on how well the model performs. Whether we use 0 or 200  amenities, the model performs the same (although it is different on the training data but this could be due to overfitting). This observation seems to suggest that amenities has no effect on the reviews_per_month.

This function below just streamlines the above process to allow me to test the effect of ameneties with different target columns

```python
def test_amenities(df, output_col, num_amenities = [0, 1, 3, 5, 10, 50, 100, 150, 200]):
    '''
    Compares the linear models from the data against one another with a different amount of amenities used.
    
    INPUT:
        df: pd.DataFrame - Dataframe containing all the features as well as the target column
        output_col: string - The name of the desired target column
        num_amenities: list<string> - A list containg the number of amenities for each iteration
    
    OUTPUT:
        best_score: dictionary - A dictionary containing the model, training data and score for the best performing model
    '''
    
    best_score = {"model": None, "X_train": None, "score": 0}
    train_scores = []
    test_scores = []

    for num in num_amenities:
        X, y = clean_data(df, output_col, num)
        lm_model, train_score, test_score, X_train = create_model(X, y)
        
        if test_score >= best_score["score"]:
            best_score["score"]= test_score
            best_score["X_train"] = X_train
            best_score["model"] = lm_model
    
        train_scores.append(round(train_score, 2))
        test_scores.append(round(test_score, 2))
    
    plt.plot(train_scores, num_amenities, label='Training Scores')
    plt.plot(test_scores, num_amenities, label='Testing Scores')
    plt.title("Predicting " + output_col) 
    plt.xlabel('R2 Score')  
    plt.ylabel('Number of amenities') 
    plt.legend()
    plt.show()
    
    return best_score
```
